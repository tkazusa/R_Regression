```python
%%R
library(zoo)
library(xts)
library(dlm)
```

```python
%%R
nonfarm <- read.csv("fredgraph.csv", header = FALSE)
```

```python
%%R 
head(nonfarm)
```

```python
%%R
nonfarm.xts <- as.xts(read.zoo(nonfarm))
```

```python
%%R
nfp <- diff(nonfarm.xts)["1939-02-01::"]
plot(nfp, minor.ticks=F, main="NFP change[thousands of persons]")
```

```python
%%R #See manual:http://cran.r-project.org/web/packages/dlm/dlm.pdf

MCMC <- 10000 #number of MCMC samples
nfp.gibbs <- dlmGibbsDIG(nfp, mod = dlmModPoly(1), a.y = 1, b.y = 1000, #nth order polynomial DLM
                         a.theta = 10, b.theta = 1000,
                         n.sample = MCMC,
                         thin = 1,
                         save.states = F)
```

```python
%%R
plot(nfp.gibbs$dV)
```

```python
%%R
plot(nfp.gibbs$dW)
```

```python
%%R
burnin <- 2000
use <-MCMC - burnin
from <- 500
plot(ergMean(nfp.gibbs$dV[-(1:burnin)], from), type="l")
```

```python
%%R
plot(ergMean(nfp.gibbs$dW[-(1:burnin)], from), type="l")
```

```python
%%R
mcmcMean(cbind(nfp.gibbs$dV[-(1:burnin)], nfp.gibbs$dW[-(1:burnin)]))
```
