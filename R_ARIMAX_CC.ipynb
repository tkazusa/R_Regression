```python
%%R
library(forecast)
library(xts)
library(ggplot2)
library(tidyr)
library(scales)
```

```python
%%R
call <- read.csv("CallCentreStats-pre2013.csv")
call.zoo <- read.zoo(call, header = TRUE, sep=",", format ="%m/%d/%Y", tz="PST8PDT")
call.xts <- as.xts(call.zoo[,"Calls.Offered"])
```

```python
%%R
#Take training data
call.train <- call.xts["::2012-11-30"]

#Set Frequency as 7days
call.train.ts <-ts(call.train, frequency=7)

#Fitting with ARIMA modeling
fit.arima <- auto.arima(call.train.ts)
```

```python
%%R
#Inference the number of call from 01/12/2012 to 31/12/2012
fc.arima <- forecast(fit.arima, h=31)

#Mean of predictive call
pred.arima <- as.vector(fc.arima$mean)
```

```python
%%R
#Take test data
call.test <- call.xts["2012-12-01::"]
```

```python
%%R
#Result from ARIMA Modele
date.test <- as.Date(index(call.xts["2012-12-1::"]))
res <- data.frame(date=date.test, pred.arima=pred.arima,
                 act = as.numeric(call.test))

#Convert Result to long format
res.m <- gather(data=res, type, value, -date)
```

```python
%%R
head(res.m)
```

```python
%%R
#Calcurate error of mean square
rmse <- sqrt(mean((res$pred.arima - res$act)^2))
rmse
```

```python
%%R
head(res)
```

```python
%%R
#Plot estimate and actual data
p <- ggplot(data=res.m, aes(x=date, y=value, color=type))
p <- p + geom_point()
p <- p + geom_line()
p <- p + scale_x_date(labels=date_format("%m/%d")) + theme(legend.position="bottom", legend.title=element_blank())
p
```
